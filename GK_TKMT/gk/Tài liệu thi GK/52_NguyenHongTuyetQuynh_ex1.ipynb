3. Viết chương trình nhập vào một dãy các số nguyên được cách nhau bởi khoảng trắng (VD: 32 8 4 6 9...) lưu vào python ở dạng mảng. Thực hiện

- Xuất mảng
- Tính tổng các phần tử của mảng
- Tìm số lớn nhất
- Tìm số bé nhất
- Sắp xếp mảng theo thứ tự giảm dần
- In ra các số chẵn trong mảng

```python
def nhap_day_so():
    day_so = input('Nhập dãy số nguyên, cách nhau bởi khoảng trắng: ')
    mang = [int(x) for x in day_so.split()]
    return mang

def xuat_mang(mang):
    print('Mảng:', mang)

def tinh_tong(mang):
    return sum(mang)

def tim_so_lon_nhat(mang):
    return max(mang)

def tim_so_be_nhat(mang):
    return min(mang)

def sap_xep_giam_dan(mang):
    return sorted(mang, reverse=True)

def in_so_chan(mang):
    so_chan = [x for x in mang if x % 2 == 0]
    print('Các số chẵn trong mảng:', so_chan)

if __name__ == '__main__':
    mang = nhap_day_so()

    xuat_mang(mang)
    print('Tổng các phần tử của mảng:', tinh_tong(mang))
    print('Số lớn nhất trong mảng:', tim_so_lon_nhat(mang))
    print('Số bé nhất trong mảng:', tim_so_be_nhat(mang))
    
    mang_giam_dan = sap_xep_giam_dan(mang)
    print('Mảng sau khi sắp xếp giảm dần:', mang_giam_dan)
    
    in_so_chan(mang)
```

5. Viết chương trình nhập vào một số nguyên dương (n>0). Cho biết n có phải là số đối xứng hay không? Biết số đối xứng là số viết xuôi hay viết ngược đều như nhau

```python
def symmetrical_num(n): 
    flag =0;
    if ( n[::-1] == n):
      flag = 1
    return flag

n = input("Nhập một số tự nhiên: ")
check = symmetrical_num(n);
if check == 1:
    print(n,'là số đối xứng')
else:
    print(n,'không phải là số đối xứng')
```
