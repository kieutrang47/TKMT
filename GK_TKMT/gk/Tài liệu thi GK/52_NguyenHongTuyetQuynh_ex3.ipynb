Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

````python
import numpy as np
````

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
arr = np.random.rand(100)
print(arr[:10])
print(arr[-5:])
for i in range(len(arr)):
    if arr[i] % 5 == 0 and arr[i] > 100:
        print(arr[i])
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
a = 0
for i in range(len(arr)):
    if arr[i] < 100:
        print(arr[i])
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
print(f"{sum(arr)}")
print(f"{arr.mean()}")
print(f"{np.median(arr)}")
print(f"{np.quantile(arr, [0.25, 0.5, 0.75])}")
print(f"{np.quantile(arr, [0.2])}")
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
for i in range(5):
    random_selection = np.random.choice(arr, size = 10, replace= False)
    print(f'Lần {i + 1}: ',random_selection)
    print()
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
dem = np.bincount(arr)
xuathien_nhieunhat= np.argmax(dem)
print('Phần tử xuất hiện nhiều nhất: ', xuathien_nhieunhat)
print('Số lần: ',dem[xuathien_nhieunhat])
````

## SEABORN



Seaborn là một thư viện hỗ trợ trực quan hóa được viết dựa trên thư viện `matplotlib`. Trong bài tập này, bạn hãy tự tìm hiểu thư viện `seaborn` và trực quan hóa dữ liệu tiền tip tại một nhà hàng.

Bạn có thể dùng đoạn mã sau để đọc dữ liệu có sẵn trong `seaborn`

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```
Hoặc bạn có thể đọc dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy lựa chọn loại đồ thị phù hợp để trả lời những câu hỏi sau:

````python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
# !pip install seaborn
````

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
Tong_tip = df.groupby("day", observed=False)["tip"].sum().reset_index()

# Barplot
sns.barplot(data=Tong_tip, x="day", y="tip")
plt.title("Tổng tiền tip theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Tổng tiền tip")
plt.show()
````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)


````python
tip_total_by_time = df.groupby("time", observed=False)["tip"].sum().reset_index()

#Barplot
sns.barplot(data=tip_total_by_time, x="time", y="tip")
plt.title("Tổng tiền tip theo thời gian dùng bữa")
plt.xlabel("Thời gian dùng bữa")
plt.ylabel("Tổng tiền tip")
plt.show()
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn


````python
df = sns.load_dataset("tips")
sns.histplot(data=df, x="total_bill",edgecolor='white', color= 'purple')
plt.title("Phân bố tổng hoá đơn")
plt.xlabel("Tổng hoá đơn (USD)")
plt.ylabel("Số lượng")
plt.show()
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
df = sns.load_dataset("tips")
sns.boxplot(data=df, y="total_bill")
plt.ylabel("Tổng hoá đơn (USD)")
plt.show()
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
df = sns.load_dataset("tips")
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("So sánh tổng hoá đơn giữa các ngày trong tuần")
plt.xlabel("Ngày trong tuần")
plt.ylabel("Tổng hoá đơn (USD)")
plt.show()
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
df = sns.load_dataset("tips")
sns.scatterplot(data=df, x="total_bill", y="tip", color='purple')
plt.title("Liên quan giữa tổng hoá đơn và tiền tip")
plt.xlabel("Tổng hoá đơn (USD)")
plt.ylabel("Tiền tip (USD)")
plt.show()
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
df = sns.load_dataset("tips")

# Sử dụng hàm jointplot để khảo sát quan hệ giữa total_bill và tip
sns.jointplot(data=df, x="total_bill", y="tip", color='red')
plt.show()
#Hai thuộc tính này có tương quan với nhau
````

### 8. Tìm hiểu và sử dụng hàm `pairplot` trong seaborn để trực quan bộ dữ liệu. Theo bạn, trong bộ dữ liệu có những thuộc tính nào liên quan với nhau không?



````python
df = sns.load_dataset("tips")
sns.pairplot(df)
plt.show()
````

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
df = sns.load_dataset("tips")
lunch = df[df["time"] == "Lunch"]["total_bill"]
dinner = df[df["time"] == "Dinner"]["total_bill"]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(lunch, bins=10, color='skyblue')
axes[0].set_title("Tổng hóa đơn - Lunch")
axes[0].set_xlabel("Tổng hóa đơn")
axes[0].set_ylabel("Số lượng")

axes[1].hist(dinner, bins=10, color='darkred')
axes[1].set_title("Tổng hóa đơn - Dinner")
axes[1].set_xlabel("Tổng hóa đơn")
axes[1].set_ylabel("Số lượng")

plt.tight_layout()
plt.show()
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
df = sns.load_dataset("tips")
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.boxplot(data=df[df["time"] == "Lunch"], x="time", y="total_bill", ax=axes[0])
axes[0].set_title("Tổng hóa đơn - Lunch")
axes[0].set_xlabel("Thời gian dùng bữa")
axes[0].set_ylabel("Tổng hóa đơn")

sns.boxplot(data=df[df["time"] == "Dinner"], x="time", y="total_bill", ax=axes[1])
axes[1].set_title("Tổng hóa đơn - Dinner")
axes[1].set_xlabel("Thời gian dùng bữa")
axes[1].set_ylabel("Tổng hóa đơn")

plt.tight_layout()
plt.show()
````
