**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
lan_Tung = 5000
mat = [1, 2, 3, 4, 5, 6]

#tung xúc sắc
mophong = np.random.choice(mat, size=lan_Tung)
# a. Xác suất để số chấm xuất hiện là 4
cham_4 = np.count_nonzero(mophong == 4) / lan_Tung
print("Xác suất để số chấm xuất hiện là 4:", cham_4)
# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
cham_lon_or_bang_4 = np.count_nonzero(mophong >= 4) / lan_Tung
print("Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4:", cham_lon_or_bang_4)
# c. Giả sử số chấm xuất hiện lớn hơn hoặc bằng 4. Tìm xác suất để mặt 6 chấm xuất hiện.
#các lần tung xúc sắc có số chấm xuất hiện lớn hơn hoặc bằng 4
mau_cham_lon_or_bang_4 = mophong[mophong >= 4]
#xác suất để mặt 6 chấm xuất hiện trong mẫu
cham_6_trong_mau = np.count_nonzero(mau_cham_lon_or_bang_4 == 6) / len(mau_cham_lon_or_bang_4)
print("Xác suất để mặt 6 chấm xuất hiện khi số chấm xuất hiện lớn hơn hoặc bằng 4:", cham_6_trong_mau)
```

**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
np.random.seed(42)

n = 50
p = 0.7
mau = 5000

# Tạo dữ liệu
a = np.random.binomial(n, p, size=mau)

# a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3
min_value = np.min(a)
max_value = np.max(a)
q1 = np.percentile(a, 25)
q2 = np.percentile(a, 50)
q3 = np.percentile(a, 75)

print("Min:", min_value)
print("Max:", max_value)
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)

# b. Vẽ đồ thị boxplot
plt.boxplot(a)
plt.title("Boxplot")
plt.show()

# c. Vẽ các đồ thị: tần số, tần suất
plt.hist(a, bins=range(min_value, max_value + 2), edgecolor='black', align="left")
plt.xticks(range(min_value, max_value + 1))
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.title("Đồ thị tần số")
plt.show()

plt.hist(a, bins=range(min_value, max_value + 2), edgecolor='black',align="left", density=True)
plt.xticks(range(min_value, max_value + 1))
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")
plt.title("Đồ thị tần suất")
plt.show()
```
