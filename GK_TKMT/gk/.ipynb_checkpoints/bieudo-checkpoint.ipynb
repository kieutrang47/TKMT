```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o', linestyle='-', color='b', label="Đường xu hướng")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ đường")
plt.legend()
plt.show()
# '-'	Đường liền nét
# '--'	Đường gạch ngang
# '-.'	Đường gạch chấm
# ':'	Đường chấm chấm
```

```python
# Bar Chart)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values, color='skyblue')
plt.xlabel("Nhóm")
plt.ylabel("Giá trị")
plt.title("Biểu đồ cột")
plt.show()
```

```python
# Biểu đồ cột ngang (Horizontal Bar Chart)
plt.barh(categories, values, color='orange')
plt.xlabel("Giá trị")
plt.ylabel("Nhóm")
plt.title("Biểu đồ cột ngang")
plt.show()
```

```python
# Biểu đồ tròn (Pie Chart)
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple'])
plt.title("Biểu đồ tròn")
plt.show()
```

```python
# Biểu đồ tán xạ (Scatter Plot)
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='blue', alpha=0.5)
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ tán xạ")
plt.show()
```

```python
# Biểu đồ Histogram (Phân phối dữ liệu)
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green')
# plt.hist(data, bins=30, color='green', edgecolor='black') có viền
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")
plt.title("Biểu đồ Histogram")
plt.show()
```

```python
# Biểu đồ hộp (Box Plot)
data = [np.random.randn(100), np.random.randn(100) + 2]

plt.boxplot(data, labels=["Nhóm 1", "Nhóm 2"])
plt.ylabel("Giá trị")
plt.title("Biểu đồ hộp")
plt.show()
```

```python
# Biểu đồ vùng (Area Plot)
x = np.arange(1, 6)
y = [5, 15, 10, 20, 25]

plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.plot(x, y, color="blue")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ vùng")
plt.show()
```

```python
# Biểu đồ lưới nhiệt (Heatmap)
import seaborn as sns

data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Biểu đồ Heatmap")
plt.show()
```

```python
# Biểu đồ đường kết hợp điểm (Stem Plot)
plt.stem(x, y, linefmt="r-", markerfmt="bo", basefmt="k-")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ Stem")
plt.show()
```
