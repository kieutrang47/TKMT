**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
'''a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi.
 Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.
'''
n, p = 10, 0.25

X = stats.binom(n, p)
sample = X.rvs(10)
print("Số câu đúng của sinh viên trong 10 lần kiểm tra là: ", sample)

# answer = np.random.choice([0, 1], size=(10, 10), p=[0.75, 0.25])
# for i in range(10):
#     print(f"Số câu đúng lần kiểm tra thứ {i+1} là: {np.sum(answer[i])}")
```

```python
'''b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm
'''
for i in range(11):
    j = 10 - i
    if i*4 + j*(-2) == 4:
        print(f"Để sinh viên được 4 điểm thì cần đúng {i} câu và sai {j} câu")

print(f"Xác suất để sinh viên được 4 điểm là: {X.pmf(4)}")
```

```python
'''c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)'''
E_X = X.mean()
Var_X = X.var()

print(f"Expected value: {E_X}")
print(f"Variance: {Var_X}")
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.

b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
m = 500
sigma = np.sqrt(16)

'''
b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. 
Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
'''
X = stats.norm(m, sigma)
sample = X.rvs(20)

# print(sample)

plt.hist(sample, bins=10, density=True, edgecolor='black')
plt.xlabel('Weight')
plt.ylabel('Density')
plt.title('Histogram of fruit weight')
plt.show()

E_X = sample.mean()
Var_X = sample.var()
Std = sample.std()

print(f"Kỳ vọng: {E_X}")
print(f"Phương sai: {Var_X}")
print(f"Độ lệch chuẩn: {Std}")


'''
a. Hãy tính tỷ lệ của mỗi loại.
'''

print('Tỷ lệ loại 1: ', (sample[sample > 505].sum() / sample.sum()).round(4))
print('Tỷ lệ loại 2: ', (sample[(sample >= 495) & (sample <= 505)].sum() / sample.sum()).round(4))
print('Tỷ lệ loại 3: ', (sample[sample < 495].sum() / sample.sum()).round(4))


'''
c. Bạn hãy mô phỏng lấy 100 trái cây trên.
 Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
'''

sample_2 = X.rvs(100)

# print(sample_2)

plt.hist(sample_2, bins=50, density=True, edgecolor='black')
plt.xlabel('Weight')
plt.ylabel('Density')
plt.title('Histogram of fruit weight')
plt.show()

E_X_2 = sample_2.mean()
Var_X_2 = sample_2.var()
Std_2 = sample_2.std()

print(f"Kỳ vọng: {E_X_2}")
print(f"Phương sai: {Var_X_2}")
print(f"Độ lệch chuẩn: {Std_2}")
```

### a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`

### Ý nghĩa: Tính xác suất để có đúng 4 lần thành công trong 10 lần thử, với xác suất thành công mỗi lần là 0.6

```python
n = 10; p = 0.6
X = stats.binom(n, p)

print("\nBINOMIAL\n")
print("Xác suất P(X <= 5): ", X.cdf(5))
print("Xác suất P(X >=4): ", 1 - X.cdf(3))
print("Xác suất P(X = 5): ", X.pmf(5))

# c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`
'''
 Ý nghĩa: Tính xác suất để có đúng 4 sự kiện xảy ra trong một khoảng thời gian cố định, 
   với tần suất trung bình là 3 sự kiện trong khoảng thời gian đó.
'''

lambda_ = 3
Y = stats.poisson(lambda_)
print("\nPOISSON\n")
print("Xác suất P(Y <= 5): ", Y.cdf(5))
print("Xác suất P(Y >= 4): ", 1 - Y.cdf(3))
print("Xác suất P(Y = 5): ", Y.pmf(5))
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
# Hàm mô phỏng tung đồng xu
def simulate_coin_toss(N, simulations=100):
    results = []
    for _ in range(simulations):
        # Mô phỏng N lần tung đồng xu
        tosses = np.random.choice([0, 1], size=N, p=[0.5, 0.5])  # 0 là Tail, 1 là Head
        head_count = np.sum(tosses)  # Đếm số lần ra Head
        results.append(head_count / N)  # Tính tỷ lệ số lần ra Head
    return np.var(results)  # Trả về phương sai của kết quả

# Các giá trị N cần mô phỏng
N_values = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
variances = []

# Tính phương sai cho mỗi giá trị N
for N in N_values:
    var = simulate_coin_toss(N)
    variances.append(var)

# Vẽ biểu đồ phương sai theo N
plt.figure(figsize=(10, 6))
plt.plot(N_values, variances, marker='o', color='r')
plt.title('Phương sai của tỷ lệ số lần tung Head theo N')
plt.xlabel('Số lần tung (N)')
plt.ylabel('Phương sai')
plt.grid(True)
plt.show()
```
