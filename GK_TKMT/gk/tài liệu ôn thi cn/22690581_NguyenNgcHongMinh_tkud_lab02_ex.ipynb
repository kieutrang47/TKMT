Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
from numpy import random
x = random.randint(1000, size=100)
print(x[:10])
print(x[-5:])
print(x[(x > 100) & (x % 5 == 0)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
from numpy import random
import numpy as np
x = random.randint(1000, size=100)
arr = x[x < 100]
print(arr.size)   
print(arr)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
from numpy import random
arr = random.randint(1000, size=100)
print(f'- Tổng: ',arr.sum())
print('- Trung bình: ', arr.mean())
print('- Trung vị: ', np.median(arr))
print('- Tứ phân vị [Q1, Q2, Q3]: ', np.quantile(arr, [0.25, 0.5, 0.75]))
print('- Phân vị 20 : ', np.quantile(arr, 0.20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
from numpy import random
import numpy as np
x = random.randint(100, size=100)
unique, counts = np.unique(x, return_counts=True)
max_count_index = np.argmax(counts)
most_frequent_element = unique[max_count_index]
most_frequent_count = counts[max_count_index]
print(f"Phần tử xuất hiện nhiều nhất: {most_frequent_element}")
print(f"Số lần xuất hiện: {most_frequent_count}")
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 6.3)
ySin = np.sin(x)
yCos = np.cos(x)
plt.plot(x, ySin, label="sin(x)", color="blue")
plt.plot(x, yCos,ls='--', label="cos(x)", color="orange")
plt.legend(loc="lower left")
plt.grid(True) 
plt.title("Đồ thị hàm sin(x) và cos(x)")
plt.show()
```

### Câu 5: Vẽ các đồ thị con

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 6.2, 100)

fig, ax = plt.subplots(2, 1, constrained_layout=True) # Tạo bố cục với 2 đồ thị con (2 dòng, 1 cột)

# Thêm tiêu đề chung cho cả biểu đồ
fig.suptitle('Đồ thị hàm sin(x) và cos(x)')

# Vẽ đồ thị sin(x) trên đồ thị con đầu tiên
ax[0].plot(x, np.sin(x), label='sin(x)')
ax[0].legend() 

# Vẽ đồ thị cos(x) trên đồ thị con thứ hai
ax[1].plot(x, np.cos(x), color='red', label='cos(x)')
ax[1].legend() 

plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 6.2, 100)

fig, ax = plt.subplots(2, 1, constrained_layout=True) # Tạo bố cục với 2 đồ thị con (2 dòng, 1 cột)

# Thêm tiêu đề chung cho cả biểu đồ
fig.suptitle('Đồ thị hàm sin(x) và cos(x)')
x = np.linspace(0, 6.2, 100)

ax[0].hist(np.sin(x), bins=50, edgecolor='black', alpha=0.7)
ax[0].set_title('Histogram')
ax[0].set_ylabel('Mật độ')

plt.xlabel('Value')
plt.ylabel('Frequence')
plt.title('Histogram')
plt.show()
# Vẽ đồ thị cos(x) trên đồ thị con thứ hai
ax[1].plot(x, np.cos(x), color='red', label='cos(x)')
ax[1].legend() 

plt.show()
```
