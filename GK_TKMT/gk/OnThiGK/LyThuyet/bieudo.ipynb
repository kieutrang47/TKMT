```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o', linestyle='-', color='b', label="Đường xu hướng")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ đường")
plt.legend()
plt.show()
# '-'	Đường liền nét
# '--'	Đường gạch ngang
# '-.'	Đường gạch chấm
# ':'	Đường chấm chấm
```

```python
# Bar Chart)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values, color='skyblue')
plt.xlabel("Nhóm")
plt.ylabel("Giá trị")
plt.title("Biểu đồ cột")
plt.show()
```

```python
# Biểu đồ cột ngang (Horizontal Bar Chart)
plt.barh(categories, values, color='orange')
plt.xlabel("Giá trị")
plt.ylabel("Nhóm")
plt.title("Biểu đồ cột ngang")
plt.show()
```

```python
# Biểu đồ tròn (Pie Chart)
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple'])
plt.title("Biểu đồ tròn")
plt.show()
```

```python
# Biểu đồ tán xạ (Scatter Plot)
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='blue', alpha=0.5)
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ tán xạ")
plt.show()
```

```python
# Biểu đồ Histogram (Phân phối dữ liệu)
data = np.random.randn(1000)

plt.hist(data, bins=30, color='green', edgecolor='black')
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")
plt.title("Biểu đồ Histogram")
plt.show()
```
