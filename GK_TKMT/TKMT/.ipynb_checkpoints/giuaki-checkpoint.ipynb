### 1.

```python
import pandas as pd

df = pd.read_csv('StudentSurvey.csv')

df.head()
```

```python
# Bỏ qua các giá trị thiếu trong cột năm
df_cleaned = df.dropna(subset=['Year'])

# Xây dựng bảng phân phối tần số
frequency_table = df_cleaned['Year'].value_counts()

frequency_table
```

### 2.

```python
# Tính tỷ lệ sinh viên đeo khuyên
piercing_rate = (df['Piercings'] > 0).mean() * 100
print(f"Tỷ lệ sinh viên đeo khuyên: {piercing_rate:.2f}%")
```

```python
# Tính tỷ lệ đeo khuyên theo giới tính
piercing_rate_by_gender = df.groupby('Gender')['Piercings'].apply(lambda x: (x > 0).mean() * 100)

print("Tỷ lệ đeo khuyên theo giới tính:")
piercing_rate_by_gender
```

### 3.

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ histogram cho ba thuộc tính
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(df['Pulse'].dropna(), bins=10, color='blue', edgecolor='black', linewidth=1.2)
plt.title('Histogram of Pulse')
plt.xlabel('Pulse')

plt.subplot(1, 3, 2)
plt.hist(df['Exercise'].dropna(), bins=10, color='green', edgecolor='black', linewidth=1.2)
plt.title('Histogram of Exercise')
plt.xlabel('Exercise (hours)')

plt.subplot(1, 3, 3)
plt.hist(df['Piercings'].dropna(), bins=10, color='red', edgecolor='black', linewidth=1.2)
plt.title('Histogram of Piercings')
plt.xlabel('Number of Piercings')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew

pulse_skew = skew(df['Pulse'].dropna())
exercise_skew = skew(df['Exercise'].dropna())
piercings_skew = skew(df['Piercings'].dropna())

print(f"Độ lệch của Pulse: {pulse_skew}")
print(f"Độ lệch của Exercise: {exercise_skew}")
print(f"Độ lệch của Piercings: {piercings_skew}")
```

### 4.

```python
# Lấy ngẫu nhiên một mẫu với cỡ mẫu là 50
sample_df = df.sample(n=50, random_state=42)

# Tính tỷ lệ nữ trong mẫu
female_ratio_sample = (sample_df['Gender'] == 'F').mean() * 100

# Tính tỷ lệ nữ từ bộ dữ liệu ban đầu
female_ratio_population = (df['Gender'] == 'F').mean() * 100

# Tính sai số so với bộ dữ liệu ban đầu
error = abs(female_ratio_sample - female_ratio_population)

print(f"Tỷ lệ nữ trong mẫu: {female_ratio_sample:.2f}%")
print(f"Tỷ lệ nữ trong bộ dữ liệu ban đầu: {female_ratio_population:.2f}%")
print(f"Sai số: {error:.2f}%")
```

```python
# Danh sách lưu các tỷ lệ nữ trong 100 lần lấy mẫu
female_ratios = []

# Lặp lại 100 lần
for i in range(100):
    sample_df = df.sample(n=50, random_state=i)
    female_ratio_sample = (sample_df['Gender'] == 'F').mean() * 100
    female_ratios.append(female_ratio_sample)

# Vẽ histogram của các giá trị tỷ lệ nữ
plt.hist(female_ratios, bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Female Ratios (100 samples, size=50)')
plt.xlabel('Female Ratio (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Hàm lặp lại việc lấy mẫu và vẽ histogram cho các cỡ mẫu khác nhau
def sample_and_plot(sample_size, num_samples=100):
    female_ratios = []
    
    # Lặp lại num_samples lần với cỡ mẫu sample_size
    for i in range(num_samples):
        sample_df = df.sample(n=sample_size, random_state=i)
        female_ratio_sample = (sample_df['Gender'] == 'F').mean() * 100
        female_ratios.append(female_ratio_sample)
    
    plt.hist(female_ratios, bins=10, color='green', edgecolor='black')
    plt.title(f'Histogram of Female Ratios ({num_samples} samples, size={sample_size})')
    plt.xlabel('Female Ratio (%)')
    plt.ylabel('Frequency')
    plt.show()

sample_and_plot(30)
sample_and_plot(100)
sample_and_plot(200)
```
