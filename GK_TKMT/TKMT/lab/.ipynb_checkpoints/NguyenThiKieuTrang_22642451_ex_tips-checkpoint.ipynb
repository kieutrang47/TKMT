# KHÁM PHÁ DỮ LIỆU

Dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy thực hiện thống kê mô tả để tìm hiểu về hành vi tip của khách hàng tại nhà hàng

```python
#: Import các thư viện cần thiết
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Hiển thị đẹp hơn
pd.set_option("display.float_format", lambda x: f"{x:.2f}")
sns.set_theme(style="whitegrid", palette="muted")
```

```python
#. Đọc dữ liệu & tiền xử lý
# =====================================================
df = pd.read_csv("/Users/trangnguyen/Downloads/TKMT_lab_03/data/tips.csv")

# Tạo cột tỷ lệ tip (% so với tổng bill)
df["tip_pct"] = df["tip"] / df["total_bill"] * 100
```

```python


# Thống kê mô tả cơ bản
n = len(df)
tip_mean = df["tip"].mean()
tip_min = df["tip"].min()
tip_max = df["tip"].max()
tip_pct_mean = df["tip_pct"].mean()
tip_pct_min = df["tip_pct"].min()
tip_pct_max = df["tip_pct"].max()

print("Số lượng hóa đơn:", n)
print(f"Tip trung bình: {tip_mean:.2f} USD")
print(f"Tip nhỏ nhất: {tip_min:.2f} USD")
print(f"Tip lớn nhất: {tip_max:.2f} USD")
print(f"Tỷ lệ tip trung bình: {tip_pct_mean:.2f}%")
print(f"Tỷ lệ tip nhỏ nhất: {tip_pct_min:.2f}%")
print(f"Tỷ lệ tip lớn nhất: {tip_pct_max:.2f}%")

# # Biểu đồ phân phối tip và tỷ lệ tip
# plt.figure(figsize=(12,5))

# # Biểu đồ tiền tip tuyệt đối
# plt.subplot(1,2,1)
# sns.histplot(df["tip"], bins=20, kde=True, color="steelblue")
# plt.title("Phân phối số tiền tip (USD)")
# plt.xlabel("Tip (USD)")
# plt.ylabel("Số lượng khách")

# # Biểu đồ tỷ lệ tip (%)
# plt.subplot(1,2,2)
# sns.histplot(df["tip_pct"], bins=20, kde=True, color="tomato")
# plt.title("Phân phối tỷ lệ tip (%)")
# plt.xlabel("Tỷ lệ tip (%)")
# plt.ylabel("Số lượng khách")

# plt.tight_layout()
# plt.show()
```

```python
# 1: đọc dữ liệu: 
#Đọc dữ liệu từ file CSV 
# df = pd.read_csv("/Users/trangnguyen/Downloads/TKMT_lab_03/data/tips.csv")
# đọc dữ liệu.
# df.head()
```

```python
# xóa cột: Unnamed
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
df.head()
```

```python
# bao nhiêu dòng, bao nhiêu thuộc tính 
print("Số dòng:", df.shape[0])
print("Số thuộc tính:", df.shape[1])
print("Danh sách thuộc tính:", df.columns.tolist())
```

```python
# thuộc tính phân loại,
categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
print("\nThuộc tính phân loại:", categorical_cols)
# thuộc tính địng lượng.
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
print("Thuộc tính định lượng:", numeric_cols)
```

```python
# Có bao nhiêu nam, bao nhiêu nữ. Tỷ lệ bao nhiêu?
sex_stats = df["sex"].value_counts()
sex_pct = df["sex"].value_counts(normalize=True).mul(100).round(2)
print("\nGiới tính:")
print(pd.concat([sex_stats, sex_pct], axis=1, keys=["Số lượng", "Tỷ lệ %"]))
```

```python
# bao nhiêu người hút thuốc. Tỷ lệ bao nhiêu?
smoker_stats = df["smoker"].value_counts()
smoker_pct = df["smoker"].value_counts(normalize=True).mul(100).round(2)
print("\nHút thuốc:")
print(pd.concat([smoker_stats, smoker_pct], axis=1, keys=["Số lượng", "Tỷ lệ %"]))
```

```python
# Ngày nào trong tuần khách hàng hay đến dùng bữa nhất. Tỷ lệ bao nhiêu?
day_stats = df["day"].value_counts()
day_pct = df["day"].value_counts(normalize=True).mul(100).round(2)
print("\nNgày trong tuần:")
print(pd.concat([day_stats, day_pct], axis=1, keys=["Số lượng", "Tỷ lệ %"]))
```

```python
# Nhà hàng phục vụ bữa ăn nào nhiều nhất (Breakfast, Lunch, Dinner)?
time_stats = df["time"].value_counts()
time_pct = df["time"].value_counts(normalize=True).mul(100).round(2)
print("\nThời gian ăn:")
print(pd.concat([time_stats, time_pct], axis=1, keys=["Số lượng", "Tỷ lệ %"]))
```

```python
# Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất là bao nhiêu người?
# Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách?
size_stats = df["size"].value_counts().sort_index()
avg_size = df["size"].mean().round(2)
print("\nNhóm khách:")
print(size_stats)
print(f"Nhóm phổ biến nhất: {size_stats.idxmax()} người")
print(f"Trung bình 1 bữa ăn có {avg_size} khách")
```

# YẾU TỐ ẢNH HƯỞNG ĐẾN TIỀN TIP



1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
5. Thống kê tổng số tiền tip nhận được theo ngày.

```python
tip_nonzero_ratio = (df["tip"] > 0).mean() * 100
print(f"Tỷ lệ hóa đơn có tip: {tip_nonzero_ratio:.2f}%")
```

```python
df["tip_pct"] = df["tip"] / df["total_bill"] * 100
tip_pct_mean = df["tip_pct"].mean()
print(f"Tỷ lệ tip trung bình trên hóa đơn: {tip_pct_mean:.2f}%")

```

```python
sex_stats = df.groupby("sex").agg(
    tip_avg=("tip", "mean"),
    tip_pct_avg=("tip_pct", "mean"),
    freq=("tip", "count")
)
print("So sánh Nam vs Nữ:")
display(sex_stats)
```

```python
smoker_stats = df.groupby("smoker").agg(
    tip_avg=("tip", "mean"),
    tip_pct_avg=("tip_pct", "mean"),
    freq=("tip", "count")
)
print("So sánh Hút thuốc vs Không hút thuốc:")
display(smoker_stats)
```

```python
tip_by_day = df.groupby("day")["tip"].sum().sort_values(ascending=False)
print("Tổng tiền tip theo ngày:")
display(tip_by_day)
```

# TRỰC QUAN DỮ LIỆU

Để dễ hình dung, bạn hãy trực quan với một số câu hỏi sau đây:

1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày
2. Vẽ đồ thị thể hiện tổng tiền tip theo bữa ăn (time)
3. Vẽ đồ thị histogram của tổng hoá đơn
4. Vẽ đồ thị boxplot của tổng hoá đơn
5. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip.
6. So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần.

```python
import matplotlib.pyplot as plt

tip_by_day = df.groupby("day")["tip"].sum()

plt.figure(figsize=(6,4))
plt.bar(tip_by_day.index, tip_by_day.values, color="pink")
plt.title("Tổng tiền tip theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Tổng tip (USD)")
plt.show()
```

```python
tip_by_time = df.groupby("time")["tip"].sum()

plt.figure(figsize=(6,4))
plt.bar(tip_by_time.index, tip_by_time.values, color="lightgreen")
plt.title("Tổng tiền tip theo bữa ăn")
plt.xlabel("Bữa ăn")
plt.ylabel("Tổng tip (USD)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["total_bill"], bins=20, color="orange", edgecolor="black")
plt.title("Phân phối tổng hoá đơn")
plt.xlabel("Tổng bill (USD)")
plt.ylabel("Số lượng")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df["total_bill"], vert=True, patch_artist=True,
            boxprops=dict(facecolor="pink"))
plt.title("Boxplot tổng hoá đơn")
plt.ylabel("Tổng bill (USD)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["total_bill"], df["tip"], alpha=0.7, color="purple")
plt.title("Quan hệ giữa tổng bill và tip")
plt.xlabel("Tổng bill (USD)")
plt.ylabel("Tip (USD)")
plt.show()
```

```python
bill_by_day = df.groupby("day")["total_bill"].sum()

plt.figure(figsize=(6,4))
plt.bar(bill_by_day.index, bill_by_day.values, color="teal")
plt.title("Tổng chi tiêu theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Tổng bill (USD)")
plt.show()
```

## Với Matplotlib

Bạn hãy trực quan với thư viện matplotlib

```python
import matplotlib.pyplot as plt

# 1. Tổng tiền tip theo ngày
tip_by_day = df.groupby("day")["tip"].sum()
plt.figure(figsize=(6,4))
plt.bar(tip_by_day.index, tip_by_day.values, color="skyblue")
plt.title("Tổng tiền tip theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Tổng tip (USD)")
plt.show()

# 2. Tổng tiền tip theo bữa ăn (time)
tip_by_time = df.groupby("time")["tip"].sum()
plt.figure(figsize=(6,4))
plt.bar(tip_by_time.index, tip_by_time.values, color="lightgreen")
plt.title("Tổng tiền tip theo bữa ăn")
plt.xlabel("Bữa ăn")
plt.ylabel("Tổng tip (USD)")
plt.show()

# 3. Histogram của tổng hoá đơn
plt.figure(figsize=(6,4))
plt.hist(df["total_bill"], bins=20, color="orange", edgecolor="black")
plt.title("Phân phối tổng hoá đơn")
plt.xlabel("Tổng bill (USD)")
plt.ylabel("Số lượng hoá đơn")
plt.show()

# 4. Boxplot của tổng hoá đơn
plt.figure(figsize=(6,4))
plt.boxplot(df["total_bill"], vert=True, patch_artist=True,
            boxprops=dict(facecolor="pink"))
plt.title("Boxplot tổng hoá đơn")
plt.ylabel("Tổng bill (USD)")
plt.show()

# 5. Mối liên quan giữa tổng hoá đơn và tiền tip
plt.figure(figsize=(6,4))
plt.scatter(df["total_bill"], df["tip"], alpha=0.7, color="purple")
plt.title("Quan hệ giữa tổng bill và tip")
plt.xlabel("Tổng bill (USD)")
plt.ylabel("Tip (USD)")
plt.show()

# 6. So sánh tổng chi tiêu giữa các ngày
bill_by_day = df.groupby("day")["total_bill"].sum()
plt.figure(figsize=(6,4))
plt.bar(bill_by_day.index, bill_by_day.values, color="teal")
plt.title("Tổng chi tiêu theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Tổng bill (USD)")
plt.show()
```

## Với Seaborn

Bạn hãy tìm hiểu và trực quan với thư viện seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")  # style trắng đẹp

fig, axes = plt.subplots(2, 3, figsize=(18,10))

# 1. Tổng tip theo ngày
sns.barplot(data=df, x="day", y="tip", estimator="sum", hue="day", 
            palette="Blues_d", legend=False, ax=axes[0,0])
axes[0,0].set_title("Tổng tiền tip theo ngày")
axes[0,0].set_xlabel("Ngày")
axes[0,0].set_ylabel("Tổng tip (USD)")

# 2. Tổng tip theo bữa ăn
sns.barplot(data=df, x="time", y="tip", estimator="sum", hue="time", 
            palette="Greens_d", legend=False, ax=axes[0,1])
axes[0,1].set_title("Tổng tiền tip theo bữa ăn")
axes[0,1].set_xlabel("Bữa ăn")
axes[0,1].set_ylabel("Tổng tip (USD)")

# 3. Histogram của tổng bill
sns.histplot(data=df, x="total_bill", bins=20, kde=True, color="orange", ax=axes[0,2])
axes[0,2].set_title("Phân phối tổng hoá đơn")
axes[0,2].set_xlabel("Tổng bill (USD)")
axes[0,2].set_ylabel("Số lượng")

# 4. Boxplot của tổng bill
sns.boxplot(data=df, y="total_bill", color="pink", ax=axes[1,0])
axes[1,0].set_title("Boxplot tổng hoá đơn")
axes[1,0].set_ylabel("Tổng bill (USD)")

# 5. Quan hệ giữa bill và tip
sns.scatterplot(data=df, x="total_bill", y="tip", hue="size", palette="viridis", ax=axes[1,1])
axes[1,1].set_title("Quan hệ giữa tổng bill và tip")
axes[1,1].set_xlabel("Tổng bill (USD)")
axes[1,1].set_ylabel("Tip (USD)")

# 6. Tổng chi tiêu theo ngày
sns.barplot(data=df, x="day", y="total_bill", estimator="sum", hue="day", 
            palette="Purples_d", legend=False, ax=axes[1,2])
axes[1,2].set_title("Tổng chi tiêu theo ngày")
axes[1,2].set_xlabel("Ngày")
axes[1,2].set_ylabel("Tổng bill (USD)")

plt.tight_layout()
plt.show()
```
