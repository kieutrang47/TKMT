## 1) Hi·ªÉn th·ªã 5 d√≤ng & B·∫£ng t·∫ßn s·ªë `Year` (b·ªè NA)

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data/StudentSurvey.csv")
display(df.head())

freq_year = df["Year"].dropna().value_counts()
freq_year
```

## 2) T·ª∑ l·ªá sinh vi√™n ƒëeo khuy√™n; so s√°nh theo gi·ªõi

```python

has_earring = (df["Piercings"] > 0)
p_hat = has_earring.mean()
print("T·ª∑ l·ªá ƒëeo khuy√™n (to√†n d·ªØ li·ªáu):", round(p_hat, 3))

ratio_by_gender = df.groupby("Gender")["Piercings"].apply(lambda s: (s>0).mean())
print("\nT·ª∑ l·ªá ƒëeo khuy√™n theo gi·ªõi:\n", ratio_by_gender)
print("\nGi·ªõi c√≥ t·ª∑ l·ªá cao h∆°n:", ratio_by_gender.idxmax())
```

## 3) Histogram: Pulse, Exercise, Piercings + nh·∫≠n x√©t l·ªách

```python

plt.figure(); plt.hist(df["Pulse"].dropna(), bins=20); plt.title("Pulse"); plt.xlabel("Pulse"); plt.ylabel("Freq"); plt.show()
plt.figure(); plt.hist(df["Exercise"].dropna(), bins=20); plt.title("Exercise"); plt.xlabel("Exercise"); plt.ylabel("Freq"); plt.show()
plt.figure(); plt.hist(df["Piercings"].dropna(), bins=20); plt.title("Piercings"); plt.xlabel("Piercings"); plt.ylabel("Freq"); plt.show()

print("Skew(Pulse)    =", df["Pulse"].dropna().skew())
print("Skew(Exercise) =", df["Exercise"].dropna().skew())
print("Skew(Piercings)=", df["Piercings"].dropna().skew())
```


## 4) `sample()`
a) L·∫•y m·∫´u n=50, t√≠nh t·ª∑ l·ªá n·ªØ; sai s·ªë so v·ªõi t·ª∑ l·ªá n·ªØ to√†n b·ªô.  
b) L·∫∑p 100 l·∫ßn (n=50), v·∫Ω histogram t·ª∑ l·ªá n·ªØ.  
c) L·∫∑p l·∫°i v·ªõi n kh√°c (20, 200) v√† nh·∫≠n x√©t.


```python

# 4a
p_pop = (df["Gender"]=="F").mean()
sample_50 = df.sample(n=50, random_state=0)
p_hat_50 = (sample_50["Gender"]=="F").mean()
print(f"T·ª∑ l·ªá n·ªØ to√†n b·ªô: {p_pop:.3f}")
print(f"T·ª∑ l·ªá n·ªØ m·∫´u (n=50): {p_hat_50:.3f}")
print(f"Sai s·ªë tuy·ªát ƒë·ªëi: {abs(p_hat_50 - p_pop):.3f}")
```

```python

# 4b
vals = []
for i in range(100):
    s = df.sample(n=50, random_state=i)
    # Gi·∫£ ƒë·ªãnh "Gender" ƒë√£ ƒë∆∞·ª£c x·ª≠ l√Ω th√†nh 'F' ho·∫∑c kh√°c (Boolean)
    vals.append((s["Gender"]=="F").mean())

vals = pd.Series(vals)
plt.figure(); plt.hist(vals, bins=15); plt.title("T·ª∑ l·ªá n·ªØ c·ªßa 100 m·∫´u (n=50)"); plt.xlabel("T·ª∑ l·ªá"); plt.ylabel("T·∫ßn su·∫•t"); plt.show()
print("Mean of sample ratios:", round(vals.mean(), 3))
print("Std of sample ratios:", round(vals.std(ddof=1), 3)) # <<< ƒê√É S·ª¨A
```

```python

# 4c
def sim(n, times=200):
    arr = []
    for i in range(times):
        # FIX: Cho ph√©p l·∫•y m·∫´u ho√†n l·∫°i (c√≥ th·ªÉ l·∫∑p l·∫°i h√†ng)
        s = df.sample(n=n, replace=True, random_state=i) 
        arr.append((s["Gender"]=="F").mean())
    return pd.Series(arr)

vals20 = sim(20, 200)
vals200 = sim(200,200)

plt.figure(); plt.hist(vals20, bins=15); plt.title("T·ª∑ l·ªá n·ªØ (n=20)"); plt.xlabel("T·ª∑ l·ªá"); plt.ylabel("T·∫ßn su·∫•t"); plt.show()
plt.figure(); plt.hist(vals200, bins=15); plt.title("T·ª∑ l·ªá n·ªØ (n=200)"); plt.xlabel("T·ª∑ l·ªá"); plt.ylabel("T·∫ßn su·∫•t"); plt.show()

print("Std(n=20) =", round(vals20.std(ddof=1), 4))
print("Std(n=200)=", round(vals200.std(ddof=1), 4))
print("Nh·∫≠n x√©t: n l·ªõn ‚Üí ph∆∞∆°ng sai nh·ªè ‚Üí ph√¢n ph·ªëi h·∫πp, b√°m g·∫ßn t·ª∑ l·ªá t·ªïng th·ªÉ.")
```
