## 2)
Kỳ vọng/Trung bình (μ): Tính giá trị trung bình của toàn bộ quần thể POP. Trong bối cảnh mô phỏng này, đây chính là Kỳ vọng lý thuyết (E[X]) của quần thể.

Độ lệch chuẩn (σ): Tính độ lệch chuẩn của toàn bộ quần thể. Hàm np.std() mặc định chia cho N, phù hợp để tính σ.

Điều kiện: Tạo ra một mảng Boolean (True/False) cho các phần tử có chiều cao từ 180 trở lên.

Đếm số lượng: True được tính là 1, False là 0. Hàm .sum() sẽ đếm tổng số người có chiều cao ≥180.

Tính Tỷ lệ: Chia tổng số người cao cho tổng kích thước quần thể (N) để có được tỷ lệ người cao.

## 3) 
Vẽ Histogram (Quần thể): Vẽ biểu đồ tần suất của POP. bins=30 chia dữ liệu thành 30 cột.

Thiết lập hiển thị: Tạo đường viền và độ trong suốt cho cột histogram.

Tạo trục x: Tạo 100 điểm cách đều nhau trong khoảng 4σ quanh μ. Đây là phạm vi điển hình để vẽ phân phối chuẩn.

Tính PDF Chuẩn: Tính giá trị của Hàm mật độ xác suất chuẩn (PDF) cho mỗi điểm x, sử dụng tham số loc (μ) và scale (σ) đã tính được ở bước 2.

Vẽ Đường cong Chuẩn: Vẽ đường cong PDF chuẩn lên biểu đồ.


## 4)
Kết luận học thuật (Trả lời Phần c)

Kết luận: Quần thể không có phân phối chuẩn.

Giải thích: Hàm np.random.randint tạo ra Phân phối Đồng đều (Uniform Distribution), nghĩa là xác suất xuất hiện của mọi chiều cao (từ 120 đến 199) là như nhau. Do đó, histogram sẽ có hình chữ nhật (hoặc gần chữ nhật), hoàn toàn khác với hình chuông (Bell Curve) của Phân phối Chuẩn mà bạn vẽ để so sánh.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Tạo quần thể mô phỏng gồm 10.000 nam thanh niên
np.random.seed(26)                # giúp tái lập kết quả
N = 10000
POP = np.random.randint(120, 200, N)  # chiều cao 120–199 cm

# 2. Tính thống kê của quần thể
trungbinh_pop = POP.mean()   # kỳ vọng E[X]
dlc_pop = POP.std()          # độ lệch chuẩn σ
tyle_cao = (POP >= 180).sum() / N

# 3. In kết quả
print("Thống kê của quần thể:")
print(f"- Kỳ vọng (mean) = {trungbinh_pop:.2f} cm")
print(f"- Độ lệch chuẩn (std) = {dlc_pop:.2f} cm")
print(f"- Tỷ lệ người cao (≥180 cm) = {tyle_cao:.2%}")

# 4. Vẽ histogram chiều cao của quần thể
plt.hist(POP, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ tần suất")
plt.title("Phân phối chiều cao của quần thể (10.000 người)")
plt.show()

# 5. Vẽ đường cong phân phối chuẩn có cùng μ, σ để so sánh
x = np.linspace(trungbinh_pop - 4*dlc_pop, trungbinh_pop + 4*dlc_pop, 100)
y = norm.pdf(x, loc=trungbinh_pop, scale=dlc_pop)
plt.plot(x, y, color='red', linewidth=2)
plt.title(f"Đường cong chuẩn μ={trungbinh_pop:.2f}, σ={dlc_pop:.2f}")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ xác suất")
plt.show()
```

```python
# 4. VẼ ĐỒ THỊ KẾT HỢP
# =======================================================

# 4.1. Vẽ Histogram (Phân phối thực nghiệm)
# density=True là RẤT QUAN TRỌNG để histogram có thể so sánh với PDF
plt.hist(POP, bins=30, edgecolor='black', alpha=0.7, density=True, label='Histogram Quần thể')

# 4.2. Vẽ Đường cong Phân phối Chuẩn (PDF Lý thuyết)
# Sử dụng trung bình (loc) và độ lệch chuẩn (scale) của quần thể
x = np.linspace(trungbinh_pop - 4*dlc_pop, trungbinh_pop + 4*dlc_pop, 100)
y = norm.pdf(x, loc=trungbinh_pop, scale=dlc_pop)
plt.plot(x, y, color='red', linewidth=2, label='Phân phối Chuẩn Lý thuyết')

# 4.3. Hoàn thiện và Hiển thị Đồ thị Chung
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ tần suất / Mật độ xác suất")
plt.title(f"So sánh Phân phối Quần thể (Uniform) và Phân phối Chuẩn\n(μ={trungbinh_pop:.2f}, σ={dlc_pop:.2f})")
plt.legend()       # Thêm chú giải để phân biệt Histogram và Đường cong
plt.grid(alpha=0.3)
plt.show()         # Chỉ gọi plt.show() một lần duy nhất
```

- Kết luận
=>Vì randint() cho giá trị phân bố đều (uniform), nên histogram không giống phân phối chuẩn → quần thể không có phân phối chuẩn.

### Bài 2: Lấy mẫu cỡ 20

Bạn hãy lấy một mẫu kích thước 20 phần từ. Tính:

a. Trung bình mẫu, và độ lệch chuẩn của mẫu

b. Tỷ lệ người cao của mẫu

c. Vẽ histogram của mẫu

d. Thử vẽ đồ thị hàm mật độ của phân phối chuẩn với tham số loc và scale bằng với giá trị trung bình và độ lệch chuẩn và chiều cao của mẫu. Bạn có nhận xét gì không?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Tạo quần thể giả định
np.random.seed(26)
POP = np.random.randint(120, 200, 10000)

# 2. Lấy mẫu ngẫu nhiên 20 phần tử
sam = np.random.choice(POP, size=20, replace=False)

# 3. Tính các thống kê mẫu
tb_mau = sam.mean()               # Trung bình mẫu
dlc_mau = sam.std()               # Độ lệch chuẩn mẫu
tyle_cao = (sam >= 180).sum() / 20  # Tỷ lệ người cao trong mẫu

# 4. In ra kết quả
print("Mẫu gồm 20 phần tử:", sam)
print(f"Trung bình mẫu = {tb_mau:.2f}")
print(f"Độ lệch chuẩn mẫu = {dlc_mau:.2f}")
print(f"Tỷ lệ người >=180cm = {tyle_cao:.2%}")

# 5. Vẽ histogram chiều cao mẫu
plt.hist(sam, bins=10, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ tần suất")
plt.title("Phân phối chiều cao của mẫu (n=20)")

# 6. Vẽ đường cong phân phối chuẩn tương ứng với mẫu
x = np.linspace(tb_mau - 4*dlc_mau, tb_mau + 4*dlc_mau, 100)
y = norm.pdf(x, loc=tb_mau, scale=dlc_mau)
plt.plot(x, y, color='red', linewidth=2)
plt.show()
```

### Bài 3: Lấy mẫu cỡ 100

Lặp lại thí nghiệm bài 2 với cỡ mẫu là 100

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Tạo quần thể giả định
np.random.seed(26)
POP = np.random.randint(120, 200, 10000)

# 2. Lấy mẫu ngẫu nhiên 100 phần tử
sam = np.random.choice(POP, size=100, replace=False)

# 3. Tính các thống kê mẫu
tb_mau = sam.mean()                 # Trung bình mẫu
dlc_mau = sam.std()                 # Độ lệch chuẩn mẫu
tyle_cao = (sam >= 180).sum() / 100 # Tỷ lệ người cao

# 4. In ra kết quả
print("Mẫu gồm 100 phần tử (hiển thị 20 phần tử đầu):", sam[:20])
print(f"Trung bình mẫu = {tb_mau:.2f}")
print(f"Độ lệch chuẩn mẫu = {dlc_mau:.2f}")
print(f"Tỷ lệ người >=180cm = {tyle_cao:.2%}")

# 5. Vẽ histogram chiều cao mẫu
plt.hist(sam, bins=20, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ tần suất")
plt.title("Phân phối chiều cao của mẫu (n=100)")

# 6. Vẽ đường cong phân phối chuẩn tương ứng với mẫu
x = np.linspace(tb_mau - 4*dlc_mau, tb_mau + 4*dlc_mau, 100) #Tạo 100 điểm để vẽ Đường cong Chuẩn, sử dụng  xˉ và s của mẫu n=100.
y = norm.pdf(x, loc=tb_mau, scale=dlc_mau)    #Tính PDF Chuẩn: Tính giá trị PDF chuẩn. Ta dùng x- và s để xem mẫu n=100 có phân phối gần chuẩn hay không.
plt.plot(x, y, color='red', linewidth=2)
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. 
- Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

  
b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. 
- Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

  
c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo quần thể mô phỏng
np.random.seed(26)
POP = np.random.randint(120, 200, 10000)

# 2. Hàm lấy mẫu nhiều lần và lưu trung bình mẫu
def lay_trung_binh_mau(POP, sample_size, so_lan=100):
    ds_trung_binh = []
    for i in range(so_lan):
        sam = np.random.choice(POP, size=sample_size, replace=False)
        ds_trung_binh.append(sam.mean())
    return ds_trung_binh

# a. 100 trung bình mẫu cỡ 20
means_20 = lay_trung_binh_mau(POP, sample_size=20)

# b. 100 trung bình mẫu cỡ 100
means_100 = lay_trung_binh_mau(POP, sample_size=100)

# 3. Vẽ histogram riêng biệt
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(means_20, bins=15, edgecolor='black', alpha=0.7)
plt.title("Phân phối trung bình mẫu (n=20)")
plt.xlabel("Giá trị trung bình")
plt.ylabel("Tần suất")

plt.subplot(1,2,2)
plt.hist(means_100, bins=15, edgecolor='black', alpha=0.7)
plt.title("Phân phối trung bình mẫu (n=100)")
plt.xlabel("Giá trị trung bình")
plt.ylabel("Tần suất")

plt.tight_layout()
plt.show()

# 4. Vẽ hai histogram trong cùng một hình để so sánh
plt.figure(figsize=(8,5))
plt.hist(means_20, bins=15, alpha=0.6, label="n = 20", edgecolor='black')
plt.hist(means_100, bins=15, alpha=0.6, label="n = 100", edgecolor='black')
plt.xlabel("Giá trị trung bình mẫu")
plt.ylabel("Tần suất")
plt.title("So sánh phân phối trung bình mẫu (n=20 và n=100)")
plt.legend()
plt.show()
```

Phần c: Phân tích Phân phối Quần thể
1. Phân tích Mô hình Sinh dữ liệu
Quần thể (POP) được tạo ra bằng hàm np.random.randint(120, 200, 10000).

Đầu vào: Hàm randint sinh ra các số nguyên ngẫu nhiên trong một phạm vi xác định.

Bản chất học thuật: Đây là một mô phỏng của Phân phối Đồng đều (Uniform Distribution), nơi xác suất xuất hiện của mọi giá trị trong khoảng [120,199] là như nhau.

2. Nhận xét Histogram (Phân phối Quần thể)
Dựa trên bản chất của dữ liệu đầu vào:

Quan sát: Biểu đồ Histogram của quần thể POP không có hình dạng chuông đối xứng (Bell Curve) mà sẽ có hình dạng gần như hình chữ nhật (flat).

Kết luận: Quần thể mô phỏng này không tuân theo Phân phối Chuẩn (Normal Distribution).

3. Giải thích Học thuật và Đối chiếu (Phần d)
Phân phối Chuẩn là phân phối xác suất quan trọng nhất trong thống kê, được định nghĩa bởi hai tham số: Kỳ vọng (μ) và Độ lệch chuẩn (σ).

Đặc trưng của Phân phối Chuẩn: Đường cong hàm mật độ xác suất (PDF) có dạng hình chuông, đối xứng hoàn hảo, với phần lớn dữ liệu tập trung xung quanh μ.

Đối chiếu: Khi bạn vẽ đường cong chuẩn (sử dụng norm.pdf) với μ và σ của quần thể, đường cong này hoàn toàn không khớp với histogram hình chữ nhật của POP.

Tóm lại: Việc phân phối không có hình chuông là hệ quả trực tiếp của việc sử dụng mô hình Uniform (randint) để mô phỏng quần thể ban đầu, chứ không phải mô hình Chuẩn.

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. 
- Vẽ histogram tỷ lệ người cao của các mẫu
  
b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500.
- Vẽ histogram tỷ lệ người cao của các mẫu

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo quần thể mô phỏng
np.random.seed(26)
POP = np.random.randint(120, 200, 10000)

# 2. Hàm lấy mẫu nhiều lần và lưu tỷ lệ người cao
def lay_ty_le_cao(POP, sample_size, so_lan=1000):
    ds_tyle = []
    for i in range(so_lan):
        sam = np.random.choice(POP, size=sample_size, replace=False)
        tyle = (sam >= 180).sum() / sample_size
        ds_tyle.append(tyle)
    return ds_tyle

# a. 1000 mẫu, mỗi mẫu 100 người
tyle_100 = lay_ty_le_cao(POP, sample_size=100)

# b. 1000 mẫu, mỗi mẫu 500 người
tyle_500 = lay_ty_le_cao(POP, sample_size=500)

# 3. Vẽ histogram riêng biệt
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(tyle_100, bins=20, edgecolor='black', alpha=0.7)
plt.title("Phân phối tỷ lệ người cao (n=100)")
plt.xlabel("Tỷ lệ người ≥180cm")
plt.ylabel("Tần suất")

plt.subplot(1,2,2)
plt.hist(tyle_500, bins=20, edgecolor='black', alpha=0.7)
plt.title("Phân phối tỷ lệ người cao (n=500)")
plt.xlabel("Tỷ lệ người ≥180cm")
plt.ylabel("Tần suất")

plt.tight_layout()
plt.show()

# 4. So sánh hai phân phối trong cùng biểu đồ
plt.figure(figsize=(8,5))
plt.hist(tyle_100, bins=20, alpha=0.6, label="n = 100", edgecolor='black')
plt.hist(tyle_500, bins=20, alpha=0.6, label="n = 500", edgecolor='black')
plt.xlabel("Tỷ lệ người cao (≥180cm)")
plt.ylabel("Tần suất")
plt.title("So sánh phân phối tỷ lệ mẫu (n=100 và n=500)")
plt.legend()
plt.show()
```

=> Kích thước mẫu càng lớn, tỷ lệ mẫu càng tiệm cận tỷ lệ thật của quần thể

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Tạo quần thể 10.000 người với chiều cao ngẫu nhiên
np.random.seed(26)
POP = np.random.randint(120, 200, 10000)

# 2. Hàm mô phỏng 1000 lần lấy mẫu và lưu phương sai
def lay_phuong_sai_mau(POP, sample_size=100, so_lan=1000):
    ds_var = []
    for i in range(so_lan):
        sam = np.random.choice(POP, size=sample_size, replace=False)
        var = sam.std() ** 2          # phương sai = (độ lệch chuẩn)^2
        ds_var.append(var)
    return ds_var

# 3. Thực hiện mô phỏng
phuong_sai_100 = lay_phuong_sai_mau(POP, sample_size=100, so_lan=1000)

# 4. Vẽ histogram phương sai
plt.figure(figsize=(8,5))
plt.hist(phuong_sai_100, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Phương sai mẫu (variance)")
plt.ylabel("Tần suất")
plt.title("Phân phối phương sai của 1000 mẫu (n = 100)")
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
# 1. Import các thư viện cần thiết
import numpy as np              # Thư viện xử lý mảng số học (tính trung bình, độ lệch chuẩn,...)
import matplotlib.pyplot as plt # Dùng để vẽ biểu đồ minh họa
from scipy.stats import norm, t # Gọi hàm thống kê từ SciPy: norm cho Z, t cho Student

# -------------------------------------------------
# 2. Khai báo dữ liệu mẫu
x = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])
# -> Đây là 9 con gà được cân, tính theo đơn vị kg.

n = len(x)          # Hàm len() trả về kích thước mẫu (n = 9)
xbar = x.mean()     # .mean() tính trung bình mẫu (x̄)
s = x.std(ddof=1)   # .std() tính độ lệch chuẩn mẫu (s)
                    # Tham số ddof=1 để chia cho (n-1) thay vì n -> đúng với thống kê mẫu

print("=== Thông tin mẫu ===")
print(f"Số mẫu n = {n}")
print(f"Trung bình mẫu (x̄) = {xbar:.3f}")
print(f"Độ lệch chuẩn mẫu (s) = {s:.3f}")

# 3. Trường hợp (a) BIẾT σ = 0.3
# -------------------------------------------------
# Khi biết σ, ta dùng phân phối chuẩn Z.
# Công thức: CI = x̄ ± z_(α/2) * (σ / √n)

sigma = 0.3                  # Độ lệch chuẩn thật của quần thể (cho sẵn)
z_crit = norm.ppf(0.975)     # norm.ppf(p): tìm giá trị Z sao cho P(Z < z_crit) = 0.975
                             # Vì 95% CI → 2.5% ở mỗi đuôi → p = 1 - α/2 = 0.975
se_z = sigma / np.sqrt(n)    # Sai số chuẩn SE = σ / √n
ci_z = (xbar - z_crit * se_z, xbar + z_crit * se_z)
                             # Tính khoảng tin cậy (trung bình ± z * SE)

print("\n=== Trường hợp (a) Biết σ = 0.3 ===")
print(f"Giá trị Z tới hạn = {z_crit:.3f}")
print(f"Sai số chuẩn SE = {se_z:.3f}")
print(f"Khoảng tin cậy 95%: ({ci_z[0]:.4f}, {ci_z[1]:.4f})")

# 4. Trường hợp (b) KHÔNG BIẾT σ
# -------------------------------------------------
# Khi không biết σ, ta phải dùng độ lệch chuẩn mẫu s và phân phối t.
# Công thức: CI = x̄ ± t_(α/2, n-1) * (s / √n)

t_crit = t.ppf(0.975, df=n-1)
# t.ppf(p, df): trả về giá trị t sao cho P(T < t_crit) = p với bậc tự do df = n - 1.
# Ở đây df = 8, p = 0.975 → độ tin cậy 95%.

se_t = s / np.sqrt(n)         # Sai số chuẩn mẫu SE = s / √n
ci_t = (xbar - t_crit * se_t, xbar + t_crit * se_t)
                              # Khoảng tin cậy 95% khi không biết σ

print("\n=== Trường hợp (b) Không biết σ ===")
print(f"Giá trị t tới hạn (df={n-1}) = {t_crit:.3f}")
print(f"Sai số chuẩn SE = {se_t:.4f}")
print(f"Khoảng tin cậy 95%: ({ci_t[0]:.4f}, {ci_t[1]:.4f})")

# 5. Vẽ biểu đồ minh họa cho hai khoảng tin cậy
# -------------------------------------------------
plt.figure(figsize=(8,5))  # Kích thước biểu đồ 8x5 inch

# Tạo trục x quanh trung bình mẫu
x_range = np.linspace(xbar - 4*s, xbar + 4*s, 200)
# np.linspace(a,b,n): tạo ra 200 giá trị đều nhau từ a đến b
# Dùng để vẽ đường cong PDF mượt

y = norm.pdf(x_range, loc=xbar, scale=s/np.sqrt(n))
# norm.pdf(x, loc, scale): tính giá trị hàm mật độ chuẩn (PDF)
# loc = trung bình, scale = độ lệch chuẩn → ở đây dùng độ lệch chuẩn của trung bình mẫu

plt.plot(x_range, y, color='blue', linewidth=2, label='Phân phối trung bình mẫu')
# Vẽ đường cong PDF màu xanh mô tả phân phối của trung bình mẫu

# Vẽ hai khoảng tin cậy
plt.axvline(ci_z[0], color='red', linestyle='--', label='CI biết σ (biên dưới)')
plt.axvline(ci_z[1], color='red', linestyle='--', label='CI biết σ (biên trên)')
plt.axvline(ci_t[0], color='green', linestyle=':', label='CI không biết σ (biên dưới)')
plt.axvline(ci_t[1], color='green', linestyle=':', label='CI không biết σ (biên trên)')

# Thêm tiêu đề và trục
plt.title("Khoảng tin cậy 95% cho trung bình cân nặng gà (kg)")
plt.xlabel("Giá trị trung bình ước lượng (kg)")
plt.ylabel("Mật độ xác suất (PDF)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# ==============================================
# Minh họa song song: khoảng tin cậy khi biết σ (Z) và khi không biết σ (t)
# ==============================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t as tdist

# ---- DỮ LIỆU MẪU ----
x = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])  # cân nặng 9 con gà (kg)
n = len(x)                 # kích thước mẫu
xbar = x.mean()            # trung bình mẫu
s = x.std(ddof=1)          # độ lệch chuẩn mẫu (ddof=1 để chia cho n-1)
sigma = 0.3                # phương sai tổng thể đã biết (dùng cho trường hợp a)
alpha = 0.05               # mức ý nghĩa (1 - alpha = 0.95 → độ tin cậy 95%)

# ---- TÍNH GIÁ TRỊ TỚI HẠN & KHOẢNG TIN CẬY ----

# (a) Trường hợp biết σ → dùng phân phối chuẩn Z
z_crit = norm.ppf(1 - alpha/2)      # tra giá trị Z_α/2
se_z = sigma / np.sqrt(n)           # sai số chuẩn SE = σ / √n
ci_z = (xbar - z_crit * se_z, xbar + z_crit * se_z)  # khoảng tin cậy 95%

# (b) Trường hợp không biết σ → dùng phân phối t
t_crit = tdist.ppf(1 - alpha/2, df=n-1)   # tra giá trị t_α/2, bậc tự do df = n-1
se_t = s / np.sqrt(n)                     # sai số chuẩn mẫu SE = s / √n
ci_t = (xbar - t_crit * se_t, xbar + t_crit * se_t)  # khoảng tin cậy 95%

# ---- TÍNH HÀM MẬT ĐỘ XÁC SUẤT (PDF) CỦA X̄ TRONG HAI TRƯỜNG HỢP ----
# Dải giá trị quanh trung bình mẫu, đủ rộng để thấy hai khoảng tin cậy
span = 5 * s
xs = np.linspace(xbar - span, xbar + span, 600)

# Trường hợp biết σ: phân phối chuẩn của trung bình mẫu
pdf_z = norm.pdf(xs, loc=xbar, scale=se_z)

# Trường hợp không biết σ: phân phối t của trung bình mẫu
scale_t = se_t
pdf_t = tdist.pdf((xs - xbar)/scale_t, df=n-1) / scale_t

# ---- VẼ BIỂU ĐỒ ----
plt.figure(figsize=(9,5))

# Vẽ đường cong mật độ cho hai trường hợp
plt.plot(xs, pdf_z, linewidth=2, label='Biết σ → x̄ ~ Chuẩn (Z)')
plt.plot(xs, pdf_t, linewidth=2, linestyle='--', label='Không biết σ → x̄ ~ t')

# Tô vùng giữa hai biên của khoảng tin cậy
plt.fill_between(xs, 0, np.where((xs>=ci_z[0])&(xs<=ci_z[1]), pdf_z, 0),
                 alpha=0.15, label='Khoảng tin cậy 95% (Z)')
plt.fill_between(xs, 0, np.where((xs>=ci_t[0])&(xs<=ci_t[1]), pdf_t, 0),
                 alpha=0.15, label='Khoảng tin cậy 95% (t)')

# Vẽ các đường thẳng thể hiện trung bình mẫu và các biên
plt.axvline(xbar, linewidth=1.5, linestyle=':', label='Trung bình mẫu x̄')
plt.axvline(ci_z[0], linestyle='--', linewidth=1, color='red')
plt.axvline(ci_z[1], linestyle='--', linewidth=1, color='red')
plt.axvline(ci_t[0], linestyle='-.', linewidth=1, color='green')
plt.axvline(ci_t[1], linestyle='-.', linewidth=1, color='green')

# Ghi tiêu đề và trục
plt.title("Khoảng tin cậy 95% cho trung bình cân nặng gà (μ): so sánh Z và t")
plt.xlabel("Giá trị trung bình ước lượng (kg)")
plt.ylabel("Mật độ xác suất (PDF)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ---- IN KẾT QUẢ RA MÀN HÌNH ----
print("=== THÔNG TIN MẪU ===")
print(f"Số quan sát n = {n}")
print(f"Trung bình mẫu x̄ = {xbar:.5f}")
print(f"Độ lệch chuẩn mẫu s = {s:.5f}")

print("\n--- Trường hợp (a) BIẾT σ = 0.3 ---")
print(f"Giá trị Z tới hạn = {z_crit:.5f}")
print(f"Sai số chuẩn SE = {se_z:.5f}")
print(f"Khoảng tin cậy 95%: ({ci_z[0]:.5f}, {ci_z[1]:.5f})")

print("\n--- Trường hợp (b) KHÔNG BIẾT σ ---")
print(f"Giá trị t tới hạn (df={n-1}) = {t_crit:.5f}")
print(f"Sai số chuẩn SE = {se_t:.5f}")
print(f"Khoảng tin cậy 95%: ({ci_t[0]:.5f}, {ci_t[1]:.5f})")
```
