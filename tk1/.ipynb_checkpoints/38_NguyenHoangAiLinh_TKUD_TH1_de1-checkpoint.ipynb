### Câu 2 (2 điểm)

Tìm hiểu các nhóm thí sinh thi các ngoại ngữ khác nhau trong bộ dữ liệu

a, (1 điểm) Có bao nhiêu nhóm mỗi nhóm có số lượng sinh bao nhiêu, tỉ lệ bao nhiêu (bao gồm cả nhóm không có thí sinh đăng ký dự thi)

b, (1 điểm) Vẽ đồ thị thể tỉ lệ các nhóm thí sinh này

#### BÀI LÀM

a, (1 điểm) Có bao nhiêu nhóm mỗi nhóm có số lượng sinh bao nhiêu, tỉ lệ bao nhiêu (bao gồm cả nhóm không có thí sinh đăng ký dự thi)

```python
df['ma_ngoai_ngu'] = df['ma_ngoai_ngu'].fillna('Khong_dang_ky')
language_groups = df['ma_ngoai_ngu'].value_counts(dropna=False) 
total_candidates = len(df)

language_groups_percentage = (language_groups / total_candidates) * 100

result = pd.DataFrame({
    'Số lượng thí sinh': language_groups,
    'Tỉ lệ (%)': language_groups_percentage
})
print(result)
```

b, (1 điểm) Vẽ đồ thị thể tỉ lệ các nhóm thí sinh này

```python
plt.figure(figsize=(10, 6))
plt.bar(language_groups.index, language_groups_percentage, color='grey')
plt.xlabel('Ngoại ngữ')
plt.ylabel('Tỉ lệ (%)')
plt.title('Tỉ lệ thí sinh theo nhóm ngoại ngữ')
plt.show()
```

### Câu 3 (2 điểm)

b, (1 điểm) Vẽ đồ thị thể tỉ lệ hai nhóm thí sinh này

Biết rằng, thí sinh thi lần đầu phải thi đủ các môn: toan, ngu_van, vat_li, hoa_hoc, sinh_hoc, lich_su, dia_ly, gdcd

```python
cac_mon_hoc = ['toan', 'ngu_van', 'vat_li', 'hoa_hoc', 'sinh_hoc', 'lich_su', 'dia_li', 'gdcd']

thi_lan_dau = df[cac_mon_hoc].notnull().all(axis=1)
so_thi_lan_dau = thi_lan_dau.sum()
so_thi_tu_do = len(df) - so_thi_lan_dau

loai_thi_sinh = pd.Series([so_thi_lan_dau, so_thi_tu_do], index=['Thí sinh thi lần đầu', 'Thí sinh thi tự do'])

ti_le_thi_sinh = (loai_thi_sinh / len(df)) * 100

plt.figure(figsize=(7, 5))
plt.bar(ti_le_thi_sinh.index, ti_le_thi_sinh, color=['pink', 'grey'])
plt.ylabel('Tỉ lệ (%)')
plt.title('Tỉ lệ Thí sinh thi lần đầu so với Thí sinh thi tự do')
plt.show()
so_thi_lan_dau, so_thi_tu_do
```

### Câu 4: (2 điểm)

a, (1 điểm) Vẽ biểu đồ thể hiện phổ điểm của các môn: toán, văn, ngoại ngữ, khtn (lý, hóa, sinh), khxh(sử, địa, gdcd)

b, (0.5 điểm) Bạn hãy đưa ra bình luận của mình

c, (0.5 điểm) Theo bạn có hai thuộc tính (điểm) nào có mối tương quan với nhau không? Bình luận

```python
df['khtn'] = df[['vat_li', 'hoa_hoc', 'sinh_hoc']].mean(axis=1)
df['khxh'] = df[['lich_su', 'dia_li', 'gdcd']].mean(axis=1)

subjects = ['toan', 'ngu_van', 'ngoai_ngu', 'khtn', 'khxh']

plt.figure(figsize=(15, 10))

for i, subject in enumerate(subjects, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[subject].dropna(), bins=20, alpha=0.7, color='black', edgecolor='white')
    plt.title(f'Phổ điểm môn {subject}')
    plt.xlabel('Điểm')
    plt.ylabel('Số lượng')

plt.tight_layout()
plt.show()
```

b, (0.5 điểm) Bạn hãy đưa ra bình luận của mình

### Câu 5: (2 điểm)

Cho k = (STT % 20) + 1

Với k tính được bạn hãy chọn một tổ hợp xét tuyển dưới đây:

- 1  -  5: A01..A05
- 6  - 10: B01..B05
- 11 - 15: C01..C05
- 16 - 20: D01..D05

a, (0.5 điểm) Cho biết có bao nhiêu thí sinh có thể xét tuyển tổ hợp mà bạn chọn?

b, (1 điểm) Vẽ phổ điểm từng môn trên cùng một đồ thị (figure), và phổ điểm tổng 3 môn trên một đồ thị (figure) khác. 

c, (0.5 điểm) Nếu lấy điểm để trao học bổng cho top 2% thí sinh mỗi tổ hợp, thì tổ hợp này cần lấy mốc điểm để xét học bổng là bao nhiêu?

```python
# a.  Cho biết có bao nhiêu thí sinh có thể xét tuyển tổ hợp mà bạn chọn?
# - 16 - 20: D01..D05
# D04: toan, ngu_van, ngoai_ngu 

df['D01'] = df['toan'].notnull() & df['ngu_van'].notnull() & df['ngoai_ngu'].notnull()
print("Số lượng thí sinh có thể xét tuyển tổ hợp D01: ", df['D01'].sum())

# b. Vẽ phổ điểm từng môn trên cùng một đồ thị (figure), và phổ điểm tổng 3 môn trên một đồ thị (figure) khác. 
# Sử dụng màu sắc để phân biệt các môn thi.
plt.figure(figsize=(10, 6))
sns.histplot(df[['toan', 'ngu_van', 'ngoai_ngu']], bins=10, kde=True, color=['blue', 'green', 'red'])
plt.title('Phổ điểm từng môn thi')
plt.xlabel('Điểm')
plt.ylabel('Số lượng')
plt.legend(['Toán', 'Ngữ văn', 'Ngoại ngữ'])
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df[['toan', 'ngu_van', 'ngoai_ngu']].sum(axis=1), bins=10, kde=True, color='purple')
plt.title('Phổ điểm tổng 3 môn thi')
plt.xlabel('Điểm')
plt.ylabel('Số lượng')
plt.show()

# c. (0.5 điểm) Nếu lấy điểm để trao học bổng cho top 2% thí sinh mỗi tổ hợp, thì tổ hợp này cần lấy mốc điểm để xét học bổng là bao nhiêu?
moc_diem = df[['toan', 'ngu_van', 'ngoai_ngu']].sum(axis=1).quantile(0.98)
print("Mốc điểm để xét học bổng cho top 2% thí sinh mỗi tổ hợp: ", moc_diem)
```
