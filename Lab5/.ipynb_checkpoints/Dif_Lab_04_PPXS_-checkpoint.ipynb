**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
# a
X = binom(10, 0.25)
print('Số câu đúng của bạn sinh viên trong 10 lần kiểm tra:', X.rvs(10))
```

```python
# b
# x = số câu đúng -> 4x - 2 * (10 - x) = 4 <=> 6x = 24 <=> x = 4
print('Xác suất sinh viên được 4 điểm là', X.pmf(4))
```

```python
# c
print('E(X):', X.mean())
print('Var(X):', X.var())
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại

b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
# a
muy = 500
std = 4
X = norm(loc=muy, scale=std)

x = [495, 505]
y = X.cdf(x)

plt.figure(figsize=(10, 3))
plt.fill_between(x=np.arange(490, 495, 0.01), 
                 y1= X.pdf(np.arange(490, 495, 0.01)),
                 color='red',
                 alpha=0.35,
                 label='$P(X < 495)$')

plt.fill_between(x=np.arange(495, 505, 0.01), 
                 y1= X.pdf(np.arange(495, 505, 0.01)),
                 color='blue',
                 alpha=0.35,
                 label='$P(495 \leq X < 505)$')

plt.fill_between(x=np.arange(505, 510, 0.01), 
                 y1= X.pdf(np.arange(505, 510, 0.01)),
                 color='orange',
                 alpha=0.35,
                 label='$P(505 \leq X)$')

plt.text(x=492.5, y=0.006, s=round(y[0], 2))
plt.text(x=499.4, y=0.02, s=round(y[1] - y[0], 2))
plt.text(x=506.5, y=0.006, s=round(1 - y[1], 2))

plt.legend(loc='upper left')
plt.show()
```

```python
# b
sam_20 = X.rvs(20)
print('Trung bình E(sam_20):', sam_20.mean())
print('Độ lệch chuẩn Std(sam_20):', sam_20.std())

plt.hist(sam_20)
plt.show()
```

```python
sam_100 = X.rvs(100)
print('Trung bình E(sam_100):', sam_100.mean())
print('Độ lệch chuẩn Std(sam_100):', sam_100.std())

plt.hist(sam_100)
plt.show()
```

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
# a
rd = np.random.choice(range(1, 7), size=5000)
print('Xác xuất để số chấm xuất hiện là 4: P(X = 4) =', sum(rd == 4) / 5000)
```

```python
# b
print('Xác suất để số chấm lớn hoặc hoặc bằng 4 là P(4 <= X) =', sum(rd >= 4) / 5000)
```

```python
# c
# P(X = 6 | X >= 4) = P(X = 6) / P(X >= 4)
print('Xác suất để số chấm xuất hiện là 6, biết số chấm xuất hiện lớn hơn hoặc bằng 4 là P(X = 6 | X >= 4) =', sum(rd == 6) / sum(rd >= 4))
```

**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
# a
X = binom.rvs(50, 0.7, size=5000)
print('Max:', X.max())
print('Min:', X.min())
print('IQR: [Q1, Q2, Q3] =', np.percentile(X, [25, 50, 75]))
```

```python
# b
plt.boxplot(X)
plt.xticks([1], ['$X \sim Binom(50, 0.7)$'])
plt.show()
```

```python
# c
x = range(0, 51)
y = [sum(X <= i) / 5000 for i in x]

plt.figure(figsize=(12, 3))
plt.subplot(121)
plt.title('Đồ thị tần suất')
sns.histplot(X, kde=True, stat='density')
plt.subplot(122)
plt.title('Đồ thị tần số')
plt.plot(x, y)
plt.show()
```

- $P(X \leq 20)$
- $P(X \ge 29.5)$
- $P(X = 23.8)$

```python
# a
print('norm.cdf(2) dùng để tính hàm mật độ xác xuất X <= 2 với X ~ norm(0, 1)')
print('norm.cdf(2, 1, 1) dùng để tính hàm mật độ xác xuất X <= 2 với X ~ norm(1, 1)')
print('norm.cdf(2, 1, 2) dùng để tính hàm mật độ xác xuất X <= 2 với X ~ norm(1, 2)')
```

**Bài 8**

Một công ty sản xuất làm ra các chi tiết máy với độ dài các chi tiết có phân phối chuẩn với kỳ vọng là 3000 mm và độ lệch chuẩn là 3 mm. Yêu cầu cho các chi tiết máy này là phải có chiều dài từ 2993 mm đến 3007 mm. Tính tỷ lệ sản phẩm lỗi của công ty

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
# e
x = np.linspace(470, 530, 10000)
y = X.pdf(x)

plt.figure(figsize=(10, 5))
sns.histplot(sam_50, kde=True, stat='density', color='b', label='50 Sample KDE')
plt.plot(x, y, label='Theoretical PDF')
plt.xlabel('Value')
plt.legend()
plt.show()
```

```python
# f Nhận xét: Số lượng mẫu ngẫu nhiên càng lớn, thì histogram có xu hướng tiệm cận với hình dạng phân phối chuẩn thực tế
sam_1000 = X.rvs(1000)

plt.figure(figsize=(10, 5))
sns.histplot(sam_1000, kde=True, stat='density', color='b', label='1000 Sample KDE')
plt.plot(x, y, label='Theoretical PDF')
plt.xlabel('Value')
plt.legend()
plt.show()
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
roll = lambda n: np.random.choice([0, 1], size=n)

N = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
means = []
vars = []
for n in N:
    sample = roll(n)
    mean = np.mean(sample)
    var = np.var(sample)
    means.append(mean)
    vars.append(var)

x = roll(100)
x_mean = np.mean(x)
x_var = np.var(x)
```

```python
plt.figure(figsize=(12, 3))
plt.subplot(121)
plt.title('Var')
plt.plot(N, vars, color='b')
plt.axhline(x_var, label='sam_100')
plt.legend()
plt.subplot(122)
plt.title('Mean')
plt.plot(N, means, color='b')
plt.axhline(x_mean, label='sam_100')
plt.legend()
plt.show()
```

```python
# Nhận xét: Khi số lượng mẫu đủ lớn, các giá trị càng gần với số liệu thống kê chuẩn
```
